## Part A

```python
# Import necessary libraries
import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
```

```python
# load in concrete data
data = pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv")
data.head()
```

```python
def network(input_shape):
    # construct the network
    model = Sequential()
    model.add(Dense(10, activation="relu", input_shape=input_shape))
    model.add(Dense(1))
    
    # compile network
    model.compile(optimizer="adam", loss="mean_squared_error")
    
    return model
```

### Split data into features and target

```python
cols = data.columns
features = data[cols[cols != "Strength"]]
target = data["Strength"]
```

```python
features.head()
```

```python
target.head()
```

### Test-Train Split with Scikit-Learn

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
def train_and_eval(X_train, X_test, y_train, y_test):
    # get input shape
    input_shape = X_train.shape[1:] 
    
    # train the network
    model = network(input_shape)
    model.fit(X_train, y_train, epochs=50)
    
    # make predictions
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    
    return mse
    
```

```python
def mse_list(X_train, X_test, y_train, y_test):
    mse_list = []
    num_iterations = 50
    
    for i in range(num_iterations):
        mse = train_and_eval(X_train, X_test, y_train, y_test)
        mse_list.append(mse)
        
    return mse_list
```

### Part A Results

```python
results = mse_list(X_train, X_test, y_train, y_test)

mean_mse = np.mean(results)
std_mse = np.std(results)
```

```python
print("Mean MSE:", mean_mse)
print("Standard Deviation of MSE:", std_mse)
```
